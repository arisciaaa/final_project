## [Сделайте один обогащенный набор данных из нескольких наборов данных](#0)

```python
# 2018/W31: Big Mac Index
data_1 = pd.read_csv('Big Mac Index.csv')

data_1 = data_1[(data_1['date'] == '2006-01-01')]
data_1.head()
```

```python
# The Big Mac Index
data_2 = pd.read_excel('big mac index_may2006.xlsx')
data_2.head()
```

```python
data_2.columns = ['name', 'local_price', 'dollar_ex', 'dollar_price', 'dollar_ppp', 'dollar_valuation']
date = np.array(['2006-05-01']*45)
data_2['date'] = date

data_2.head()
```

```python
concatenated = pd.concat([data_1, data_2]).sort_values(by = 'name')
concatenated
```

Удалю колонки, которые не совпали в двух датасетах (в одном из отсутсвовали)

```python
concatenated.isna().sum()
```

```python
concatenated.columns
```

```python
concatenated.drop(['iso_a3', 'currency_code', 'USD_raw', 'EUR_raw', 'GBP_raw', 'JPY_raw', 'CNY_raw',
       'GDP_dollar', 'adj_price', 'USD_adjusted', 'EUR_adjusted',
       'GBP_adjusted', 'JPY_adjusted', 'CNY_adjusted', 'dollar_ppp',
       'dollar_valuation'], axis = 1)
```
